```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
premier_league = pd.read_csv("premier_league.csv")
premier_league.head()
```

```python
city_data = premier_league[premier_league.team == "Manchester City"] 
```

```python
y_data = 'wins'
city_data[y_data]
```

```python
plt.figure(figsize=(15,8))
plt.title("Broj golova Man City-ja u sezonama 2006-2018",fontdict={"fontweight":"bold"})
plt.xlabel("Sezona")
plt.ylabel("Broj golova")
plt.grid(True)
plt.plot(city_data.season,city_data.goals,marker = "o")
plt.xticks(rotation=30)
plt.yticks(rotation=30)
plt.savefig("Broj golova Man City-ja u sezonama 2006-2018",dpi=72)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("Broj golova Man City-ja u sezonama 2006-2018",fontdict={"fontweight":"bold"})
plt.xlabel("Sezona")
plt.ylabel("Broj golova")
plt.grid(True)
plt.bar(city_data.season,city_data.goals)
plt.xticks(rotation=60)
plt.savefig("Broj golova Man City-ja u sezonama 2006-2018",dpi=300)
plt.show()
```

```python
def plot_sezona(y_label,y_data,title,bar=False,x_label="Sezona",x_data="season"):
    plt.figure(figsize=(15,8))
    plt.title(title,fontdict={"fontweight":"bold"})
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    if bar:
        plt.bar(city_data[x_data],city_data[y_data])
        title+='_bar'
    else:
        plt.plot(city_data[x_data],city_data[y_data],marker = "o")
    plt.xticks(rotation=60)
    plt.savefig(title,dpi=300)
    plt.show()

```

```python
plot_sezona("Broj pobeda","wins","Broj pobeda Man City-ja u sezonama 2006-2018")
```

```python
plot_sezona("Broj pobeda","wins","Broj pobeda Man City-ja u sezonama 2006-2018",bar=True)
```

```python
interesantni_podaci=[
  {
        'title':"Broj golova Man City-ja u sezonama 2006-2018",
        'y_data':"goals",
        'y_label':"Broj golova",
    },
    {
        'title':"Broj pobeda Man City-ja u sezonama 2006-2018",
        'y_data':"wins",
        'y_label':"Broj pobeda",
    }   
]
```

```python
for podaci in interesantni_podaci:
    plot_sezona(**podaci)
    plot_sezona(**podaci,bar=True)
```

```python
results = pd.read_csv("results_12seasons.csv")
team_summary = premier_league.groupby("team").sum()
```

```python
def display_gen_stats():
    general_stats = ["wins","losses","goals","total_yel_card","total_red_card"]
    return (team_summary[general_stats])

def display_team_play_stats():
    team_play_stats = ["total_pass","total_through_ball","total_long_balls",
    "backward_pass","total_cross", "corner_taken"]
    return (team_summary[team_play_stats])

def display_att_stats():
    attacking_stats = ["total_scoring_att","ontarget_scoring_att","hit_woodwork","att_hd_goal",
    "att_pen_goal", "att_freekick_goal","att_ibox_goal","att_obox_goal", "goal_fastbreak", "total_offside"]
    return (team_summary[attacking_stats])

def display_def_stats():
    defensive_stats = ['clean_sheet','goals_conceded', 'saves', 'outfielder_block', 'interception',
    'total_tackle', 'last_man_tackle', 'total_clearance', 'head_clearance','own_goals', 'penalty_conceded',
    'pen_goals_conceded']
    return (team_summary[defensive_stats])

def display_other_stats():
    other_stats = ['touches', 'big_chance_missed','clearance_off_line', 'dispossessed', 'penalty_save',
    'total_high_claim', 'punches']
    return (team_summary[other_stats])
```

```python
city_results = results.loc[(results["home_team"] == "Manchester City") | (results["away_team"] == "Manchester City")]
city_results
```

```python
city_results['result'].value_counts()
```

```python
city_results_home = results.loc[(results["home_team"] == "Manchester City")]
city_results_away =  results.loc[(results["away_team"] == "Manchester City")]
```

```python
res_home = city_results_home['result'].value_counts()
```

```python
res_away = city_results_away['result'].value_counts()
```

```python
city_wins = res_home['H'] + res_away['A']
city_wins
city_draws = res_home['D'] + res_away['D']
city_draws

city_loses = res_home['A'] + res_away['H']
city_loses
```

```python
labels = ["wins","draws","losses"]
colors = ["darkgreen","yellow","darkred"]
plt.figure(figsize=(12,9))
plt.title("Manchester City overall away performances (2006-2018)",fontdict={"fontweight":"bold","fontsize":15})
plt.pie([city_wins,city_draws,city_loses],labels=labels,colors=colors,autopct="%1.2f%%",explode=(0.05,0.05,0.05))
plt.savefig("Manchester City overall away performances (2006-2018)",dpi=300)
plt.show()
```
